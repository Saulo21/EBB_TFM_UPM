```python
df_train = df_train.drop(columns=['Vehicle Class'])
df_test = df_test.drop(columns=['Vehicle Class'])

df_train
```

```python
from sklearn.model_selection import train_test_split

X = df_train.drop(columns=['Smog'])
Y = df_train['Smog']
```

```python
#!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTENC

sm = SMOTENC(
    categorical_features=[0,3,4],
    random_state=33, 
    k_neighbors=5,
    sampling_strategy='not majority'
)
x_smoted, y_smoted = sm.fit_resample(X, Y)
```

```python

x_train, x_test, y_train, y_test = train_test_split(x_smoted, y_smoted, test_size=0.20, random_state=33)
```

```python
from sklearn import preprocessing

scaler = preprocessing.RobustScaler().fit(x_train)
x_train_smoted = scaler.transform(x_train)
x_test_smoted = scaler.transform(x_test)

x_benchmark = scaler.transform(df_test)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import cross_val_score, KFold

from scipy.stats import sem
from sklearn.metrics import make_scorer, f1_score

from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import (
    f1_score, make_scorer, classification_report, log_loss, precision_score,
    confusion_matrix
)
from sklearn.model_selection import KFold, cross_val_score
from scipy.stats import sem
import numpy as np
import mlflow
import mlflow.sklearn
import pandas as pd

# Definir el modelo base y el modelo de Bagging
dt = DecisionTreeClassifier(
    class_weight="balanced",
    criterion='gini',
    max_depth=9, 
    max_leaf_nodes=None, 
    min_samples_leaf=1, 
    min_samples_split=2, 
    splitter='random'
)

bagging_optimized_model = BaggingClassifier(
    estimator=dt, 
    n_estimators=200,
    bootstrap=False, 
    bootstrap_features=False
)

# Definir el scorer
scorer_bagging = make_scorer(f1_score, average='weighted', zero_division=0)

# Entrenar el modelo
bagging_optimized_model.fit(x_train_smoted, y_train)

# Evaluar el modelo con validación cruzada
cv = KFold(50, shuffle=True, random_state=33)
scores = cross_val_score(bagging_optimized_model, x_train_smoted, y_train, cv=cv, scoring=scorer_bagging)

print("Mean score: ", np.mean(scores), "( +/- ", sem(scores), ")")

# Configurar MLflow
remote_server_uri = "http://127.0.0.1:8080"  # Reemplaza esto con la URI de tu servidor MLflow
mlflow.set_tracking_uri(remote_server_uri)
mlflow.set_experiment("bagging_experiment")

# Iniciar una nueva ejecución en MLflow
with mlflow.start_run(run_name="Bagging_Example"):
    # Registrar parámetros del modelo
    mlflow.log_param("n_estimators", bagging_optimized_model.n_estimators)
    mlflow.log_param("bootstrap", bagging_optimized_model.bootstrap)
    mlflow.log_param("bootstrap_features", bagging_optimized_model.bootstrap_features)
    mlflow.log_param("base_estimator", str(bagging_optimized_model.estimator))
    mlflow.log_param("scorer", "f1_score (weighted)")
    
    # Evaluación del modelo
    y_pred = bagging_optimized_model.predict(x_test_smoted)
    accuracy = np.mean(y_pred == y_test)
    print("Test Accuracy:", accuracy)
    
    # Registrar métricas del modelo
    mlflow.log_metric("accuracy", accuracy)
    
    # Calcular el F1-score
    f1_score_result = classification_report(y_test, y_pred, output_dict=True)
    print("F1-score:\n", classification_report(y_test, y_pred))
    
    # Registrar métricas adicionales
    mlflow.log_metric("f1_score", f1_score_result["weighted avg"]["f1-score"])
    
    # Calcular la "pérdida" utilizando log_loss
    y_pred_proba = bagging_optimized_model.predict_proba(x_test_smoted)
    loss = log_loss(y_test, y_pred_proba)
    print("Test Log Loss:", loss)
    
    # Registrar la pérdida en MLflow
    mlflow.log_metric("loss", loss)
    
    # Calcular y registrar precision_score
    precision = precision_score(y_test, y_pred, average='weighted')
    mlflow.log_metric("precision_score", precision)
    
    # Calcular matriz de confusión
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:\n", cm)
    
    # Calcular y registrar métricas derivadas para cada clase
    for i in range(len(cm)):
        tn = cm.sum() - cm[i, :].sum() - cm[:, i].sum() + cm[i, i]
        fp = cm[:, i].sum() - cm[i, i]
        fn = cm[i, :].sum() - cm[i, i]
        tp = cm[i, i]
        mlflow.log_metric(f"false_negatives_class_{i}", fn)
        mlflow.log_metric(f"false_positives_class_{i}", fp)
        mlflow.log_metric(f"true_negatives_class_{i}", tn)
        mlflow.log_metric(f"true_positives_class_{i}", tp)
    
    # Registrar el modelo
    mlflow.sklearn.log_model(bagging_optimized_model, "model")

    # Guardar el run_id
    run_id = mlflow.active_run().info.run_id
    print(f"Run ID: {run_id}")

    # Crear el DataFrame de evaluación
    eval_data = pd.DataFrame(x_test_smoted, columns=[f'feature_{i}' for i in range(x_test_smoted.shape[1])])
    eval_data["label"] = y_test
    eval_data["predictions"] = y_pred

    # Registrar las predicciones en el mismo run
    pd_dataset = mlflow.data.from_pandas(eval_data, predictions="predictions", targets="label")
    mlflow.log_input(pd_dataset, context="testing")

print("Modelo registrado exitosamente en MLflow.")
```
